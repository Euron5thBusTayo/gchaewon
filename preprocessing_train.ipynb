train 데이터 이어붙이기 22/9 ~ 23/8까지 월별 서울시 버스 승하차 데이터

```python
path = '/content/drive/MyDrive/Euron/프로젝트/'


train_files = ['bus_22_9.csv', 'bus_22_10.csv', 'bus_22_11.csv', 'bus_22_12.csv', 'bus_23_1.csv', 'bus_23_2.csv',
        'bus_23_3.csv', 'bus_23_4.csv', 'bus_23_5.csv', 'bus_23_6.csv', 'bus_23_7.csv', 'bus_23_8.csv']

train_df = [pd.read_csv(path+'train/'+f, encoding='cp949') for f in train_files]

train = pd.concat(train_df)

print(train.shape)
train.head(1)
```

## 2. 전처리

### 1) 불필요한 피처 삭제

등록일자, 교통수단타입코드, 버스정류장ARS번호, 표준버스정류장ID, 노선명

```python
train.columns
```

```python
train = train.drop(['등록일자', '교통수단타입코드', '버스정류장ARS번호', '표준버스정류장ID', '노선명'], axis =1)
train.columns
```

### 2) 심야 시간 10시 ~ 5시 데이터 삭제

```python
hours_to_drop = ['22시승차총승객수', '22시하차총승객수', '23시승차총승객수', '23시하차총승객수',
                 '00시승차총승객수', '00시하차총승객수', '1시승차총승객수', '1시하차총승객수',
                 '2시승차총승객수', '2시하차총승객수', '3시승차총승객수', '3시하차총승객수',
                 '4시승차총승객수', '4시하차총승객수', '5시승차총승객수', '5시하차총승객수']

train = train.drop(columns=hours_to_drop, axis=1)
```

```python
train.head(1)
```

### 3) 심야버스 제거

```python
train = train[train['교통수단타입명'] != '서울심야버스']
count = train[train['교통수단타입명'] == '서울심야버스'].shape[0]
print(f'서울심야버스인 행의 개수: {count}')
```

### 4) 역 고유 번호 피처 추가

역명 피처의 (00000) 5자리 번호를 역 고유 값 피처 역번호로 추가하기

```python
train['역번호'] = train['역명'].str.extract(r'\((\d+)\)')
```

```python
train.head(1)
```

### 5) 피처명 변경
- month(월 단위)
- bus_route_id(표준 버스 정류장 ID)
- bus_id(노선 번호)
- station_name(역명)
- X~Y ride(1시간 단위 승차 총 승객 수)
- X~Y takeoff(1시간 단위 하차 총 승객 수)
- bus_type(교통 수단 타입명)
- bus_station_ID (역 고유 번호)

```python
train.columns
```

```python
columns={'사용년월':'month', '노선번호': 'bus_id',
         '역명':'station_name', '교통수단타입명':'type', '역번호': 'bus_station_ID'}

train.rename(columns=columns, inplace=True)
```

```python
# 승하차 칼럼명 변경
for hour in range(6, 22):
    ride_column = f'{hour}시승차총승객수'
    new_ride_column = f'{hour}~{hour+1}_ride'
    train.rename(columns={ride_column: new_ride_column}, inplace=True)

    takeoff_column = f'{hour}시하차총승객수'
    new_takeoff_column = f'{hour}~{hour+1}_takeoff'
    train.rename(columns={takeoff_column: new_takeoff_column}, inplace=True)
```

```python
train.head(1)
```

```python
train.columns
```

### 6) 2시간씩 묶기

```python
ride_columns = [col for col in train.columns if '_ride' in col]
takeoff_columns = [col for col in train.columns if '_takeoff' in col]

# 새로운 칼럼 이름
new_ride_columns = ['6~8_ride', '8~10_ride', '10~12_ride', '12~14_ride', '14~16_ride', '16~18_ride', '18~20_ride', '20~22_ride']
new_takeoff_columns = ['6~8_takeoff', '8~10_takeoff', '10~12_takeoff', '12~14_takeoff', '14~16_takeoff', '16~18_takeoff', '18~20_takeoff', '20~22_takeoff']

# 각 시간대에 대해 ride와 takeoff 칼럼을 합쳐서 새로운 칼럼 생성
for i in range(len(new_ride_columns)):
    ride_column_names = ride_columns[i:i+2]
    takeoff_column_names = takeoff_columns[i:i+2]

    train[new_ride_columns[i]] = train[ride_column_names].sum(axis=1)
    train[new_takeoff_columns[i]] = train[takeoff_column_names].sum(axis=1)

# 기존의 ride와 takeoff 칼럼 삭제
train.drop(columns=ride_columns + takeoff_columns, inplace=True)
```

## 7) 날씨 데이터 (강수량, 평균 기온 칼럼 추가)

```python
rain = pd.read_csv(path+'data/rain.csv', encoding='cp949')
rain = rain[8:20] # 22년 9월 ~ 23년 8월 데이터만 추출
rain = rain['강수량(mm)']
rain.reset_index(drop=True, inplace=True)
rain
```

```python
temp = pd.read_csv(path+'data/temp.csv', encoding='cp949')
temp = temp[8:20] # 22년 9월 ~ 23년 8월 데이터만 추출
temp = temp['평균기온(℃)']
temp.reset_index(drop=True, inplace=True)
temp
```

```python
month = [202209, 202210, 202211, 202212, 202301, 202302, 202303, 202304, 202305, 202306, 202307,202308]
month = pd.DataFrame(month, columns = ['month'])

weather = pd.concat([month, rain, temp], axis=1)
weather.rename(columns={'강수량(mm)': 'rain', '평균기온(℃)': 'temp'}, inplace=True)
weather
```

```python
train = pd.merge(train, weather, on='month', how='left')
train
```
