## 2. 전처리

### 1) 불필요한 피처 삭제

등록일자, 교통수단타입코드, 버스정류장ARS번호, 노선명

```python
train.columns
```

```python
train = train.drop(['등록일자', '교통수단타입코드', '버스정류장ARS번호','노선명'], axis =1)
train.columns
```

### 2) 퇴근 시간 이후 (20시 ~ 5시) 데이터 삭제

```python
hours_to_drop = ['20시승차총승객수', '20시하차총승객수', '21시승차총승객수', '21시하차총승객수',
                 '22시승차총승객수', '22시하차총승객수', '23시승차총승객수', '23시하차총승객수',
                 '00시승차총승객수', '00시하차총승객수', '1시승차총승객수', '1시하차총승객수',
                 '2시승차총승객수', '2시하차총승객수', '3시승차총승객수', '3시하차총승객수',
                 '4시승차총승객수', '4시하차총승객수', '5시승차총승객수', '5시하차총승객수']

train = train.drop(columns=hours_to_drop, axis=1)
```

```python
train.head(1)
```

### 3) 심야버스 제거

```python
train = train[train['교통수단타입명'] != '서울심야버스']
count = train[train['교통수단타입명'] == '서울심야버스'].shape[0]
print(f'서울심야버스인 행의 개수: {count}')
```

### 4) 역 고유 번호 피처 추가

역명 피처의 (00000) 5자리 번호를 역 고유 값 피처 역번호로 추가하기

```python
train['역번호'] = train['역명'].str.extract(r'\((\d+)\)')
train.drop('역명', axis=1, inplace=True)
train.tail(1)
```

### 5) 피처명 변경
- month(월 단위)
- bus_station_SID (표준버스정류장ID)
- bus_station_ID (역명 뒤 다섯자리)
- bus_ID(노선 번호)
- type(교통 수단 타입명)
- X~Y ride(1시간 단위 승차 총 승객 수)
- X~Y takeoff(1시간 단위 하차 총 승객 수)

```python
train.columns
```

```python
columns={'사용년월':'month', '노선번호': 'bus_ID', '표준버스정류장ID': 'bus_station_ID',
        '교통수단타입명':'type', '역번호': 'bus_station_ID', '표준버스정류장ID': 'bus_station_SID'}

train.rename(columns=columns, inplace=True)
```

```python
# 승하차 칼럼명 변경
for hour in range(6, 20):
    ride_column = f'{hour}시승차총승객수'
    new_ride_column = f'{hour}~{hour+1}_ride'
    train.rename(columns={ride_column: new_ride_column}, inplace=True)

    takeoff_column = f'{hour}시하차총승객수'
    new_takeoff_column = f'{hour}~{hour+1}_takeoff'
    train.rename(columns={takeoff_column: new_takeoff_column}, inplace=True)
```

```python
train.head(1)
```

```python
train.columns
```

### 6) 2시간씩 묶기

```python
ride_columns = [col for col in train.columns if '_ride' in col]
takeoff_columns = [col for col in train.columns if '_takeoff' in col]

# 새로운 칼럼 이름
new_ride_columns = ['6~8_ride', '8~10_ride', '10~12_ride', '12~14_ride', '14~16_ride', '16~18_ride', '18~20_ride']
new_takeoff_columns = ['6~8_takeoff', '8~10_takeoff', '10~12_takeoff', '12~14_takeoff', '14~16_takeoff', '16~18_takeoff', '18~20_takeoff']

# 각 시간대에 대해 ride와 takeoff 칼럼을 합쳐서 새로운 칼럼 생성
for i in range(len(new_ride_columns)):
    ride_column_names = ride_columns[i:i+2]
    takeoff_column_names = takeoff_columns[i:i+2]

    train[new_ride_columns[i]] = train[ride_column_names].sum(axis=1)
    train[new_takeoff_columns[i]] = train[takeoff_column_names].sum(axis=1)

columns_to_keep = ['month', 'bus_station_SID', 'bus_station_ID', 'bus_ID', 'type']
train.drop(columns=ride_columns + takeoff_columns, inplace=True)
train = train[columns_to_keep + new_ride_columns + new_takeoff_columns]
```


2) type을 서울간선버스 - 0, 서울지선버스-1, 서울마을버스-2, 서울순환버스-3, 서울광역버스-4으로 카테고리화
