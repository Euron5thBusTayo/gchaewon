## 2. 전처리

### 1) 불필요한 피처 삭제

등록일자, 교통수단타입코드, 버스정류장ARS번호, 노선명

```python
test.columns
```

```python
test = test.drop(['등록일자', '교통수단타입코드', '버스정류장ARS번호', '노선명'], axis =1)
test.columns
```

### 2) 퇴근 시간 이후 (20시 ~ 5시) 데이터 삭제

```python
hours_to_drop = ['20시승차총승객수', '20시하차총승객수', '21시승차총승객수', '21시하차총승객수',
                 '22시승차총승객수', '22시하차총승객수', '23시승차총승객수', '23시하차총승객수',
                 '00시승차총승객수', '00시하차총승객수', '1시승차총승객수', '1시하차총승객수',
                 '2시승차총승객수', '2시하차총승객수', '3시승차총승객수', '3시하차총승객수',
                 '4시승차총승객수', '4시하차총승객수', '5시승차총승객수', '5시하차총승객수']

test = test.drop(columns=hours_to_drop, axis=1)
```

```python
test.head(1)
```

### 3) 심야버스 제거

```python
test = test[test['교통수단타입명'] != '서울심야버스']
count = test[test['교통수단타입명'] == '서울심야버스'].shape[0]
print(f'서울심야버스인 행의 개수: {count}')
```

### 4) 역 고유번호 피처 추가 & 역명 피처 제거

역명 뒤 다섯자리

```python
test['역번호'] = test['역명'].str.extract(r'\((\d+)\)')
test.drop('역명', axis=1, inplace=True)
test.tail(1)
```

### 5) 피처명 변경
- month(월 단위)
- bus_station_SID (표준버스정류장ID)
- bus_station_ID (역명 뒤 다섯자리)
- bus_ID(노선 번호)
- type(교통 수단 타입명)
- X~Y ride(1시간 단위 승차 총 승객 수)
- X~Y takeoff(1시간 단위 하차 총 승객 수)

```python
test.columns
```

```python
columns={'사용년월':'month', '노선번호': 'bus_ID', '표준버스정류장ID': 'bus_station_ID',
        '교통수단타입명':'type', '역번호': 'bus_station_ID', '표준버스정류장ID': 'bus_station_SID'}

test.rename(columns=columns, inplace=True)
```

```python
# 승하차 칼럼명 변경
for hour in range(6, 20):
    ride_column = f'{hour}시승차총승객수'
    new_ride_column = f'{hour}~{hour+1}_ride'
    test.rename(columns={ride_column: new_ride_column}, inplace=True)

    takeoff_column = f'{hour}시하차총승객수'
    new_takeoff_column = f'{hour}~{hour+1}_takeoff'
    test.rename(columns={takeoff_column: new_takeoff_column}, inplace=True)
```

```python
test.columns
```

### 6) 2시간씩 묶기

```python
ride_columns = [col for col in test.columns if '_ride' in col]
takeoff_columns = [col for col in test.columns if '_takeoff' in col]

# 새로운 칼럼 이름
new_ride_columns = ['6~8_ride', '8~10_ride', '10~12_ride', '12~14_ride', '14~16_ride', '16~18_ride', '18~20_ride']
new_takeoff_columns = ['6~8_takeoff', '8~10_takeoff', '10~12_takeoff', '12~14_takeoff', '14~16_takeoff', '16~18_takeoff', '18~20_takeoff']

# 각 시간대에 대해 ride와 takeoff 칼럼을 합쳐서 새로운 칼럼 생성
for i in range(len(new_ride_columns)):
    ride_column_names = ride_columns[i:i+2]
    takeoff_column_names = takeoff_columns[i:i+2]

    test[new_ride_columns[i]] = test[ride_column_names].sum(axis=1)
    test[new_takeoff_columns[i]] = test[takeoff_column_names].sum(axis=1)

columns_to_keep = ['month', 'bus_station_SID', 'bus_station_ID', 'bus_ID', 'type']
test.drop(columns=ride_columns + takeoff_columns, inplace=True)
test = test[columns_to_keep + new_ride_columns + new_takeoff_columns]
```

```python
test.columns
```

```python
test.head(2)
```

### 7) 카테고리형 피처 처리

1) bus_id(기존 노선번호) 대신 노선 정보에서 routeID (노선 고유 ID)로 대체하기  

https://data.seoul.go.kr/dataList/OA-15262/S/1/datasetView.do


```python
route_info = pd.read_csv(path+'data/route_info.csv', encoding='utf-8')
route_info.rename(columns={'노선명': 'bus_ID', 'ROUTEID': 'route_ID'}, inplace=True)
route_info
```

```python
test = pd.merge(test, route_info, on='bus_ID', how='left')
test
```

```python
null_routes_df = test[test['route_ID'].isnull()][['bus_ID']].drop_duplicates()
print("route_ID 값이 null인 bus_ID:")
print(null_routes_df)
```

버스 노선 고유 아이디 (route_ID)와 매핑되지 못한 데이터 처리

```python
null_routes = {'01': 100100001, '8001': 106000004, '9403': 100100390,
               '은평01': 111900002, '마포14': 113900006, '6649': 114000001, '6654': 100100322}
null_routes = pd.DataFrame.from_dict(null_routes, orient='index', columns=['route_ID']).reset_index()
null_routes.columns = ['bus_ID', 'route_ID']

null_routes
```

```python
test = pd.merge(test, null_routes, on='bus_ID', how='left')
test
```

```python
test['route_ID'] = test['route_ID_x'].fillna(test['route_ID_y'])
test.drop(['route_ID_x', 'route_ID_y'], axis=1, inplace=True)
test.insert(4, 'route_ID', test.pop('route_ID'))
```


2) type을 서울간선버스 - 0, 서울지선버스-1, 서울마을버스-2, 서울광역버스 -3으로 카테고리화

```python
type_mapping = {
    '서울간선버스': 0,
    '서울지선버스': 1,
    '서울마을버스': 2,
    '서울광역버스': 3
}
test['type'].replace(type_mapping, inplace=True)

test.head(3)
```
